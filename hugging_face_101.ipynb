```python
from transformers import pipeline
```

```python
sentiment_analyser = pipeline("sentiment-analysis")
```

```python
sentence = "I feel great today!"
```

```python
sentiment_analyser(sentence)
```

```python
neg_sentence = "There was a devastating accident, a lot were injured!"
```

```python
sentiment_analyser(neg_sentence)
```
